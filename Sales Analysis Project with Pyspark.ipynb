```python
from pyspark.sql import SparkSession
Spark = SparkSession.builder\
    .appName('salesAnalysis')\
        .getOrCreate()
```

Creating scheme

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

sales_schema= StructType([

StructField('product_id', IntegerType(), True),
StructField('customer_id', StringType(), True),
StructField('order_date', DateType(), True),
StructField('location', StringType(), True),
StructField('source_order', StringType(), True)

]
                   )
```

```python
menu_schema= StructType([
    StructField('product_id', IntegerType(), True),
    StructField('product_name', StringType(), True),
    StructField('price', FloatType(), True)


]
                        )
```

Reading files

```python
df_menu= spark.read.format('csv')\
    .option('header', 'true')\
        .option('inferschema', 'true')\
            .schema(menu_schema)\
        .load('/Volumes/myfiles/default/salesanalysisproject1/menu.csv.txt')
```

```python
df_sales= spark.read.format('csv')\
    .option('Header', 'true')\
        .option('inferschema', 'true')\
            .schema(sales_schema)\
            .load('/Volumes/myfiles/default/salesanalysisproject1/sales.csv.txt')
```

Display Tables

```python
df_menu.show()
df_sales.show()
```

Creating separate columns for Month, year, Quarter

```python
from pyspark.sql.types import *
df_sales = df_sales.withColumn('Order_Month', month(df_sales.order_date))
df_sales.show()
```

```python
df_sales= df_sales.withColumn('Order_quarter', quarter(df_sales.order_date))
df_sales.show()
```

```python
df_sales = df_sales.withColumn('Order_year', year(df_sales.order_date))
df_sales.show()
```

Display full Table

```python
display(df_sales) 
```

## Use **Cases**

Display Total amount spent by each customer

```python
df_Tot_amt_bycustomer= df_sales.join(df_menu, 'product_id').groupBy('customer_id').agg({'price':'sum'}).alias('Total_sum').orderBy('customer_id')

display(df_Tot_amt_bycustomer)
```

Display Total amount spent on each category

```python
df_tot_amt_percat= df_sales.join(df_menu, 'product_id').groupBy('product_name').agg({'price': 'sum'}).alias('tol_sum').orderBy('product_name')

df_tot_amt_percat.show()
display(df_tot_amt_percat)
```

Display total sales in each month

```python
df_monthly_sale= df_sales.join(df_menu, 'product_id').groupBy("Order_Month").agg({'price': "sum"}).alias('sale per month').orderBy('Order_Month')

df_monthly_sale.show()
display(df_monthly_sale)
```

Display sales per year

```python
yearly_sale = df_sales.join(df_menu, 'product_id').groupBy('Order_year').agg({'price':'sum'}).alias('Sales per year').orderBy('Order_year')

yearly_sale.show()
display(yearly_sale) 
```

Display sales per quarter

```python
sales_per_quarter= df_sales.join(df_menu, 'product_id').groupBy("Order_quarter").agg({'price':'sum'}).alias("salesper qtr").orderBy("Order_quarter")

sales_per_quarter.show()
display(sales_per_quarter)
```

Display total no of order per each category

```python
from pyspark.sql.functions import *
total_ord= (
    df_sales.join(df_menu, 'product_id')
    .groupBy('product_id', 
             'product_name')
    .agg(count('product_id').alias('total_o'))
    .orderBy('total_o', ascending= 0)
)
total_ord.show()
display(total_ord)
```

Display Top 5 ordered item

```python
top5_ord_item = df_sales.join(df_menu, 'product_id').groupBy('product_id', 'product_name').agg(count('product_id').alias('Top 5 ordered items')).orderBy('Top 5 ordered items', ascending=0).drop('product_id').limit(5)

top5_ord_item.show()
display(top5_ord_item)


```

Display fequency of customers visited

```python
cust_visit_freq = df_sales.filter(df_sales.source_order== 'Restaurant').groupBy('customer_id').agg(countDistinct('order_date').alias('cnt')).orderBy('cnt')

display(cust_visit_freq)
```

Display total sales by country

```python
total_sale_bycountry = df_sales.join(df_menu, 'product_id').groupBy('location').agg(sum('price').alias('total sale by country'))

display(total_sale_bycountry)



```

Display total sales by source-order

```python
sale_by_source_order= df_sales.join(df_menu, 'product_id').groupBy('source_order').agg(sum('price').alias('total sales as per source order'))

display(sale_by_source_order)
```
